```python
def to_float(x):
    x = x.replace('$', '')
    x = x.replace(',', '')
    x = float(x)
    return x

nat34['GrAppv'] = nat34['GrAppv'].apply(to_float)
nat34['SBA_Appv'] = nat34['SBA_Appv'].apply(to_float)

nat34['SBA_ratio'] = nat34['SBA_Appv'] / nat34['GrAppv']
```
